```julia
using Revise

# Loading the finite element package Amaru
using Amaru
```

```julia
# Mesh generation
blocks = [
    Block2D( [0 0; 2 0.3], nx=5, ny=1, shape=QUAD4),
]

msh = Mesh(blocks, verbose=true);
```

```julia
mplot(msh, field="point-id", pointmarkers=true)
```

```julia
# Finite element modeling

materials = [
    MaterialBind(:solids, InElasticSolid(E=2e3, nu=0.) ),
]

loggers = [
    #Logger(:node, 1)  # monitor id 1
    Logger(:node, :(x==2 && y==0)), 

    GroupLogger(:ips, :(y<0.15)),
    
    GroupLogger(:nodes, :(y==0)) 
]

domain = Domain(msh, materials, loggers, thickness=20);
```

```julia
# Define the boundary conditions
bcs = [
    BC(:node, :(x==0), :(ux=0, uy=0) ),
    BC(:node, :(x==2 && y==0), :(fx=-10) ),
    BC(:node, :(x==2 && y==0.3), :(fx=+10) ),
]

# Perform the finite element analysis
solve!(domain, bcs, nincs=1, verbose=true)

save(domain, "dom1.vtk")
```

```julia
mplot(domain, field="syy", warpscale=1.0, elev=30, azim=30)
```

```julia
tab = loggers[2].book.tables[end]
```

```julia
using PyPlot
plot(tab[:y], tab[:eyy], "o")
xlabel("y")
ylabel("syy")
title("Syy stress at lower integration points")
```
